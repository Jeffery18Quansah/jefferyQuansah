#### Data Preprocessing

The null values of the described data set can be filled using the mean values of their said series.This can be obtained from the described data set above.

```python
#Selecting the mean values
mean = list(redWine.describe().loc['mean'])
```

```python
mean
```

```python
#filling the null values in the series using the fillna method
redWine['fixed acidity'] = redWine['fixed acidity'].fillna(mean[0])
redWine['volatile acidity'] = redWine['volatile acidity'].fillna(mean[1])
redWine['chlorides'] = redWine['chlorides'].fillna(mean[4])
redWine['free sulfur dioxide'] = redWine['free sulfur dioxide'].fillna(mean[5])
redWine['density'] = redWine['density'].fillna(mean[7])
redWine['quality'] = redWine['quality'].fillna(mean[11])
```

The series are reassigned with their complete set of values


```python
#showing that all columns should be void of null values
redWine.info()
```

```python
#checking for duplicated values
redWine.duplicated()
```

```python
#removing duplicated values
redWine.drop_duplicates()
```

It is noticed that the number of columns and rows increased and decreased respectively signifying that a change took place after duplicated values were removed


### Graphs

#### A relationship between alcohol and quality is represented graphically

```python
pp.figure(figsize=[11,11])
pp.bar(redWine['quality'],redWine['alcohol'])
pp.xlabel('Quality')
pp.ylabel('Alcohol')
pp.show()
```

Alcohol  between ranges 12 and 14 have higher quality

```python
#getting a correlation model
redWine.corr()
```

```python
#visualising the correlation model
import statsmodels.api as sm
sm.graphics.plot_corr(redWine.corr(),xnames=list(redWine.corr().columns))
```

The provided graph demonstrates correlations using different tones of blue and brown. Blue shades represent lower correlation, while brown tones indicate a much higher correlation. The analysis reveals that the alcohol column exhibits a significantly higher correlation with the quality column compared to other columns. By further examining the graph, it is possible to make predictions about the quality of wines produced.
